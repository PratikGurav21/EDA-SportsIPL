# Visualization

```python
plt.figure(figsize=(12,6))
data = data2.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
plt.show()
```

Mumbai Indians are the winners in most of the matches & Rising Pune Supergients are the lost more matches.

```python
top_players = data2.player_of_match.value_counts()[:10]
#sns.barplot(x="day", y="total_bill", data=df)
fig, ax = plt.subplots(figsize=(14,9))
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="deep");
plt.show()
```

CH Gayle is the most Successful player in all match winners

```python
temp_df = data2.drop_duplicates(subset=['season'], keep='last')[['season', 'winner']].reset_index(drop=True)
temp_df
```

```python
temp_series = data2.toss_decision.value_counts()
labels = (np.array(temp_series.index))
sizes = (np.array((temp_series / temp_series.sum())*100))
colors = ['Red', 'blue']
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Toss decision percentage")
plt.show()
```

Almost 55% of the toss decisions are made to field first. Now let us see how this decision varied over time.

```python
plt.figure(figsize=(12,6))
sns.countplot(x='season', hue='toss_decision', data=data2)
plt.xticks(rotation='vertical')
plt.show()
```

It seems during the initial years, teams wanted to bat first. Voila.! Look at the 2016 season, most of the toss decisions are to field first.

```python
#most common dismissal types in IPL.
plt.figure(figsize=(12,6))
sns.countplot(x='dismissal_kind', data=data1)
plt.xticks(rotation='vertical')
plt.show()
```
